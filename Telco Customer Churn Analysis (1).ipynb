#converted 0 and 1 values of senior citizen to yes/no to make it easier to understand

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a crosstab for Churn vs SeniorCitizen
churn_data = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Convert to percentage (row-wise)
churn_percent = churn_data.div(churn_data.sum(axis=1), axis=0) * 100

# Plot the stacked bar chart
ax = churn_percent.plot(kind='bar', stacked=True, figsize=(4,4), colormap='viridis')

#  Add percentage labels
for i, total in enumerate(churn_percent.values):
    cumulative = 0
    for j, val in enumerate(total):
        label_position = cumulative + val / 2
        ax.text(i, label_position, f'{val:.1f}%', ha='center', va='center', color='white', fontsize=9)
        cumulative += val

# Step 5: Format plot
plt.title("Churn by SeniorCitizen (in %)")
plt.xlabel("SeniorCitizen (0 = No, 1 = Yes)")
plt.ylabel("Percentage")
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```

```python

# List of columns to plot
features = ['PhoneService', 'MultipleLines', 'InternetService',
            'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
            'TechSupport', 'StreamingTV', 'StreamingMovies']

# Create subplots - 3 rows x 3 columns
fig, axes = plt.subplots(3, 3, figsize=(15, 12))  # Adjust size as needed
fig.suptitle("Countplots of Service Features", fontsize=16)

# Flatten axes for easier iteration
axes = axes.flatten()

# Loop through columns and create countplots
for i, feature in enumerate(features):
    sns.countplot(x=feature, hue=feature, data=df, ax=axes[i], palette='Set2', legend=False)
    axes[i].set_title(f"{feature}", fontsize=12)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=30)


# Remove any unused subplots if features < total subplots
for j in range(len(features), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout(rect=[0, 0, 1, 0.97])  # Adjust layout to fit title
plt.show()
```
